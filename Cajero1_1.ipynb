```python
import numpy as np
import time

# Usuarios creados
usuarios = [
    {
        "account": "123456",
        "password": "1234",
        "balance": np.random.randint(5.0, 1000.0)
    },
    {
        "account": "234567",
        "password": "2345",
        "balance": np.random.randint(5.0, 1000.0)
    },
    {
        "account": "345678",
        "password": "3456",
        "balance": np.random.randint(5.0, 1000.0)
    },
    {
        "account": "456789",
        "password": "4567",
        "balance": np.random.randint(5.0, 1000.0)
    },
    {
        "account": "567890",
        "password": "5678",
        "balance": np.random.randint(5.0, 1000.0)
    }
]

# Aqui las funciones las cuales le asigna el balance ramdom

def create_user(_account_number, _password):
    return {
            "account": _account_number,
            "password": _password,
            "balance": np.random.randint(5.0, 1000.0)
        }

def get_user(_account_number):
    users_account = [x["account"] for x in usuarios]
    if (_account_number in users_account):
        current_user_index = None
        for x in range(len(usuarios)):
            if (usuarios[x]["account"] == _account_number):
                current_user_index = x
        return usuarios[current_user_index], current_user_index
    else:
        return None, None

def login_user(_account_number):
    user, _ = get_user(_account_number)

    if (user):
        password = input("Password: ")
        if (password == user["password"]):
            return user
        else:
            print("Contraseña incorrecta")
            return None
    else:
        print("Usuario no encontrado")
        return None

def validate_user():
    account_number = input("Account Number: ")
    user = login_user(account_number)
    if (user):
        return user
    else:
        password = input("Password: ")
        new_user = create_user(account_number, password)
        usuarios.append(new_user)
        return new_user

def show_menu():
    print("""
        Ingresa la opción que deseas realizar:

        1) Consultar saldo

        2) Depositar cantidad de dinero

        3) Retirar dinero

        4) Transferir dinero

        5) Salir

        ---------------------------------
    """)

def show_balance(_user):
    print(f"Tu saldo actual es: ${_user['balance']:.2f}")

def deposit_money(_user):
    amount = float(input("Ingresa la cantidad a depositar: "))
    if amount <= 0:
        print("La cantidad a depositar debe ser mayor que cero.")
        return
    _user["balance"] += amount
    print(f"""
        Deposito de ${amount:.2f} realizado a las {time.strftime('%X')}

        Saldo actual: ${_user['balance']:.2f}
        =================================
    """)

def withdraw_money(_user):
    amount = float(input("Ingresa la cantidad a retirar: "))
    if amount <= 0:
        print("La cantidad a retirar debe ser mayor que cero.")
        return
    if amount > _user["balance"]:
        print("Saldo insuficiente.")
        return
    _user["balance"] -= amount
    print(f"""
        Retiro de ${amount:.2f} realizado a las {time.strftime('%X')}

        Saldo actual: ${_user['balance']:.2f}
        =================================
    """)

def transfer_money(_user):
    amount = float(input("Ingresa la cantidad a transferir: "))
    if amount <= 0:
        print("La cantidad a transferir debe ser mayor que cero.")
        return
    if amount > _user["balance"]:
        print("Saldo insuficiente.")
        return

    recipient_account = input("Ingresa el número de cuenta del destinatario: ")
    recipient, recipient_index = get_user(recipient_account)

    if recipient is None:
        print("Cuenta de destinatario no encontrada.")
        return

    _user["balance"] -= amount
    usuarios[recipient_index]["balance"] += amount

    print(f"""
        Transferencia de ${amount:.2f} realizada a las {time.strftime('%X')}
        Destinatario: {recipient_account}

        Tu saldo actual: ${_user['balance']:.2f}
        =================================
    """)

def main():
    end = False
    current_logged_user = None

    while not end:
        while not current_logged_user:
            current_logged_user = validate_user()

        show_menu()
        option = input("Selecciona una opción: ")

        if option == "1":
            show_balance(current_logged_user)
        elif option == "2":
            deposit_money(current_logged_user)
        elif option == "3":
            withdraw_money(current_logged_user)
        elif option == "4":
            transfer_money(current_logged_user)
        elif option == "5":
            print("Gracias por usar nuestros servicios. ¡Hasta pronto!")
            end = True
        else:
            print("Opción no válida. Por favor, selecciona una opción del menú.")

        if not end:
            input("Presiona Enter para continuar...")

if __name__ == "__main__":
    main()
```
